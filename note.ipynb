    Grafico de diagrama de barras de los estados de los pozos

```python
estado= datos['ESTADO'].values
inac=0
pro=0
hl=0
cerr=0
aban=0
ent=0
bal=0
sup=0
sw=0
caja=[]
for i in estado:
    if i =='Inactivo':
        inac+=1
        caja.append(5)
    elif i =='Producción':
        pro+=1
        caja.append(6)
    elif i=='HL':
        hl+=1
        caja.append(4)
    elif i=='Cerrado':
        cerr+=1
        caja.append(2)
    elif i=='Enterrado':
        ent+=1
        caja.append(3)
    elif i=='Abandonado':
        aban+=1
        caja.append(0)
    elif i=='Suspendido':
        sup+=1
        caja.append(7)
    elif i=='SWAB':
        sw+=1
        caja.append(8)
    else:
        bal+=1
        caja.append(1)
est=['Inactivo','Producción','Herramienta Local','Cerrado','Abandonado','Enterrado','Balancin','Suspendido','SWAB']
num=[inac,pro,hl,cerr,aban,ent,bal,sup,sw]
plt.rcParams["figure.figsize"] = (15, 15)
fig, ax = plt.subplots()
ax.set_ylabel('N de Pozos')
ax.set_title('Cantidad de Pozos')
plt.bar(est, num)
plt.show()
```

Diagrama de cajas y bigotes de los estados de los pozos

```python
caja.sort()
fig = plt.figure(figsize=(10, 7))
plt.boxplot(caja)
plt.show()
```
